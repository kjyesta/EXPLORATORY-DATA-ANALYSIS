### Cummulative sum of the people effected by the viral disease caused by mosquito bites?

```python
Week_list = df1['MMWR Week'].tolist()
#print(Week_list)
```

```python
MMWR_Week_list = []
for week in Week_list:
    if week not in MMWR_Week_list:
        MMWR_Week_list.append(week)
#print(len(MMWR_Week_list))
#print(MMWR_Week_list)
```

```python
malarial_group = df1.groupby('MMWR Week')
malarial_list = malarial_group['Malaria, Cum 2017'].count().tolist()
malarial_series = pd.Series(malarial_list)
#print(len(malarial_list))
#print(malarial_list)
malarial_cumsum = malarial_series.cumsum().tolist()
#print(malarial_cumsum)
```

```python
dengue_group = df2.groupby('MMWRWeek')
dengue_list = dengue_group['Dengue Virus Infection, Dengue§, Cum 2017'].count().tolist()
dengue_series = pd.Series(dengue_list)
#print(len(dengue_list))
#print(dengue_list)
dengue_cumsum=dengue_series.cumsum().tolist()
#print(dengue_cumsum)
```

```python
X,=plt.plot(MMWR_Week_list, malarial_cumsum, color='g', label="malaria")
Y,=plt.plot(MMWR_Week_list, dengue_cumsum, color='orange', label="dengue" )
plt.xlabel('MMWR_Week_list')
plt.ylabel('Cummulative Sum')
plt.title('Viral Diseases Caused By Mosquitoes in the year 2017')
plt.legend([X, Y], ['malaria', 'dengue'])
plt.show()
```

### Cummulative sum of the people effected by the viral disease caused by microparasites?

```python
Legionellosis_group = df1.groupby('MMWR Week')
Legionellosis_list = Legionellosis_group['Legionellosis, Cum 2017'].count().tolist()
Legionellosis_series = pd.Series(Legionellosis_list)
#print(len(Legionellosis_list))
#print(Legionellosis_list)
Legionellosis_cumsum=Legionellosis_series.cumsum().tolist()
#print(Legionellosis_cumsum)
```

```python
Cryptosporidiosis_group = df2.groupby('MMWRWeek')
Cryptosporidiosis_list = Cryptosporidiosis_group['Cryptosporidiosis, Cum 2017'].count().tolist()
Cryptosporidiosis_series = pd.Series(Cryptosporidiosis_list)
#print(len(Cryptosporidiosis_list))
#print(Cryptosporidiosis_list)
Cryptosporidiosis_cumsum=Cryptosporidiosis_series.cumsum().tolist()
#print(Cryptosporidiosis_cumsum)
```

```python
X,= plt.plot(MMWR_Week_list, Legionellosis_cumsum, color='g')
Y,= plt.plot(MMWR_Week_list, Cryptosporidiosis_cumsum, color='orange')
plt.xlabel('MMWR_Week_list')
plt.ylabel('Cummulative Sum')
plt.title('Viral Diseases Caused by Microparasites in 2017')
plt.legend([X, Y], ['Legionellosis', 'Cryptosporidiosis'])
plt.show()
```

### Which disease is more effective in US in Malaria/Dengue?

```python
area_list = df1[' Reporting Area'].tolist()
#print(area_list)
```

```python
uniqueList = []

for letter in area_list:
    if letter not in uniqueList:
        uniqueList.append(letter)
#print(len(uniqueList))
#print(uniqueList)
```

```python
m_area_group = df1.groupby(' Reporting Area')
new_malarial_list = m_area_group['Malaria, Cum 2017'].count().tolist()
#print(len(new_malarial_list))
#print(new_malarial_list)
```

```python
d_area_group = df2.groupby('Reporting Area')
new_dengue_list = d_area_group['Dengue Virus Infection, Dengue§, Cum 2017'].count().tolist()
#print(len(new_dengue_list))
#print(new_dengue_list)
```

```python
df = pd.DataFrame({'new_malarial_list': new_malarial_list,
                  'new_dengue_list': new_dengue_list}, index=uniqueList)
ax = df.plot.bar(rot=90,figsize=[15,5])
```
